defining a little function to plot a map with basemap

for more about `contourf` function, see [matplotlib.pyplot doc](http://matplotlib.org/api/pyplot_api.html)

```python
def plot_field(m, X, lat, lon, vmin, vmax, step, \
               cmap=plt.get_cmap('jet'), ax=False, title=False, grid=False):
    if not ax:
        f, ax = plt.subplots(figsize=(8, (X.shape[0] / float(X.shape[1])) * 8))
    m.ax = ax
    im = m.contourf(lons, lats, X, np.arange(vmin, vmax+step, step), \
                   latlon=True, cmap=cmap, extend='both', ax=ax)
    m.drawcoastlines()
    m.drawcountries()
    if grid:
        m.drawmeridians(np.arange(0, 360, 30), labels=[0,0,0,1])
        m.drawparallels(np.arange(-80, 80,20), labels=[1,0,0,0])
    m.colorbar(im)
    if title:
        ax.set_title(title)
```

|Keyword   |Description|
|----------|:----------|
|lat_ts|latitude of true scale. Optional for stereographic, cylindrical equal area and mercator projections. default is lat_0 for stereographic projection. default is 0 for mercator and cylindrical equal area projections.|

#### other parameters:
|Keyword   |Description|
|----------|:----------|
|resolution|resolution of boundary database to use. Can be c (crude), l (low), i (intermediate), h (high), f (full) or None. ### _`groupby( )`_  method and How to calculate climatologies


```python
dset850 = dset.sel(level=850)
```

** notice!** the parameter in _`mean( )`_, is the name or index of a axis or dimension

```python
clim850 = dset850.groupby('time.month').mean('time')
```

```python
clim850
```

```python
from calendar import month_name
```

```python
# `figsize=( , )`, unit is inches
f, axes = plt.subplots(nrows=4, ncols=3, figsize=(14,10))
f.subplots_adjust(hspace=0.1, wspace=0.1)
axes = axes.flatten()
for i, month in enumerate(range(1,13)):
    ax = axes[i]
    plot_field(m, clim850['air'][i,:,:], \
               lats, lons, \
               -30, 30, 5, \
               ax=ax, title=month_name[month])
```

```python
f, axes = plt.subplots(nrows=1, ncols=2, figsize=(9,5))
f.subplots_adjust(wspace=0.1)
axes = axes.flatten()
for i, month in enumerate([1, 7]):
    ax = axes[i]
    plot_field(m, clim850['air'][month-1,:,:], \
               lats, lons, \
               -30, 30, 5, \
               ax=ax, title=month_name[month])
```

### Using _`apply( )`_ to calculate anomolies:
+ see pandas 'split-apply-combine' machinery
+ here, the algorithm is good, he use the 'broadcasting' machinery in numpy

**Notice!** 'broadcasting' is very useful in anomoly calculation!

```python
def demean(x):
    return x - x.sel(time=slice('1980-1-1', '1999-12-31')).mean('time')
```

>**My note:** groupby( ) method split the data into some parts, in this case each part is a ndarray with dimensions as [time, lat, lon], each part or ndarray as a whole is applied a function defined above, and here the 'broadcasting' machinery is used, 'broadcasting' along 'time' axis.

```python
air_anoms = dset850.groupby('time.month').apply(demean)
```

```python
air_anoms
```

```python
plot_field(m, air_anoms['air'].sel(time='1989-1-1'), lats, lons, \
               -2, 2, 0.1, title='anomaly', \
           cmap=plt.get_cmap('RdBu_r'))
```

**Summary on data type conversion:**
+ from .nc file to Dataset, _`open_dataset( )`_
+ from Dataset to DataArray, _`[var]`_
+ from DataArray to Dataset, _`to_dataset( )`_
+ from Dataset to .nc, _`to_netcdf( )`_

```python
anoms = air_anoms['air'].to_dataset()
anoms.to_netcdf('./air.central_asia.anomaly.1989.nc')
```

```python
d = {}
d['time'] = ('time',time)
d['latitudes'] = ('latitudes',lat)
d['longitudes'] = ('longitudes', lon)
d['var'] = (['time','latitudes','longitudes'], arr)
```

```python
dset_from_dict = xray.Dataset(d)
```

```python
dset_from_dict.longitudes.attrs['units'] = 'degrees_east'
dset_from_dict.latitudes.attrs['units'] = 'degrees_north'
```

```python
dset_from_dict.to_netcdf('./dset_from_dict.nc')
```
